```python
sales=pd.read_csv('sales.csv')
sales
```

```python
sales.isnull().sum()
```

```python
sales['Order Date'] = pd.to_datetime(sales['Order Date'], format="%m/%d/%Y", errors='coerce')
sales
```

```python
sales['Delivery Date'] = pd.to_datetime(sales['Delivery Date'], format="%m/%d/%Y", errors='coerce').dt.date
sales
```

```python
sales['Delivery Date']=sales['Delivery Date'].fillna(0)
sales
```

```python
sales.isnull().sum()
```

```python
sales=sales.rename(columns={'Order Date':'Order_date'})
sales=sales.rename(columns={'Delivery Date':'Delivery_date'})
```

```python
sales.info()
```

```python
#check duplicates in sales data
print(sales.duplicated().sum())
```

```python
#product data cleaning
product=pd.read_csv("Products.csv")
product
```

```python
product.isnull().sum()
```

```python
product.info()
```

```python
#convert string into float, remove '$' and ',' , remove spaces using str.strip
product['Unit Cost USD'] = product['Unit Cost USD'].str.replace('$', '').str.replace(",", "").str.strip().astype(float)
product['Unit Price USD'] = product['Unit Price USD'].str.replace('$', '').str.replace(",", "").str.strip().astype(float)
product.head(10)
```

```python
product.info()
```

```python
product=product.rename(columns={'Product Name':'Product_name'})
product=product.rename(columns={'Unit Cost USD':'Unit_cost_usd'})
product=product.rename(columns={'Unit Price USD':'Unit_price_usd'})
product
```
